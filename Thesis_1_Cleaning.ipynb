## Protecting data

```python
#cleaning data
customers.columns #['Cus_ID', 'Cus_OID', 'Cus_Name', 'Cus_Agency', 'Cus_Zone']

#verifying the cosistency of the Ids
customers.pivot_table(index='Cus_ID',aggfunc='size').to_frame().sort_values(by=0,ascending='False') #no repetition of IDS
print('no repetition of IDs')
```

```python
#cleanning Customers dataset
customers_no_name=customers.drop(columns=['Cus_Name']) #removing full names 

#treating new categories for agencies column
agengy_list=list(customers_no_name.Cus_Agency.unique())
new_agency_categories=['A','B','C','D','E','F','G','H','I']
zip_agency=zip(agengy_list,new_agency_categories)
dict_agency=dict(list(zip_agency))

customers_no_name=customers_no_name.assign(Agency=customers_no_name['Cus_Agency'].map(dict_agency)) #creating a categories of Agencies

#treating new categories for zones column
zone_list=list(customers_no_name.Cus_Zone.unique())
new_zone_categories=list(np.arange(len(zone_list)))
zip_zone=zip(zone_list,new_zone_categories)
dict_zone=dict(list(zip_zone))

customers_no_name=customers_no_name.assign(Zone=customers_no_name['Cus_Zone'].map(dict_zone)) #creating a categories of zones

#cleaned data
data_customers=customers_no_name.drop(columns=['Cus_Agency','Cus_Zone'])
data_customers.head(10)
```

```python
#integrating new columns
data_clients_complementary=pd.read_csv(address+'data_clients_complementary.csv',sep=';') #has additional columns like sex and birth date and civil state

data_customers_merge=pd.merge(data_customers,data_clients_complementary,left_on='Cus_ID',right_on='cli_id',how='left')
data_customers_merge=data_customers_merge.rename({'ent_sexo': 'Sex', 'ent_fechanacimiento': 'Birth','ent_estadocivil':'Marital_Status'}, axis='columns')

data_customers_merge.Birth=pd.to_datetime(data_customers_merge.Birth, format='%d/%m/%Y')
data_customers_merge=data_customers_merge.assign(age=(datetime.strptime('2017-12-01', '%Y-%m-%d')-data_customers_merge.Birth).astype('timedelta64[D]')/365)
data_customers=data_customers_merge

data_customers=data_customers[(~data_customers.Cus_OID.isna())&(data_customers.Cus_OID!=0) ] #removing nulls and 0
data_customers=data_customers[~data_customers.age.isna()]
```

```python
#cleanning the age column 
data_customers  #4536 original number of rows

df=data_customers[~(data_customers.age<18) &~(data_customers.age>90)] #1184 possible outliers
mean=df.age.mean()
stdv=df.age.std()
np.random.seed(123)
n=data_customers[(data_customers.age<18)|(data_customers.age>90)].age.count()
randages=list(np.random.normal(mean, stdv, n))
ids=list(data_customers[(data_customers.age<18)|(data_customers.age>90)].Cus_ID)

df=pd.DataFrame(data=zip(ids,randages),columns=['Cus_ID','age'])

data_customers.loc[data_customers.Cus_ID.isin(df.Cus_ID),['age']]=df.age.values #replacing with new ages
data_customers=data_customers.drop(columns=['Birth'])
```

```python
#Cleaning account datase.
accounts_cleaning=accounts.assign(duration=accounts.Acc_Mat_date-accounts.Acc_Start_date) #creating the variable duration such that variable acc_plan will be removed.

list_terms=list(accounts_cleaning.Acc_Rith.unique()) #renaming terms in a new understandable category
new_terms=['Week','Day','2 Weeks','Month','Month and a half','2 Weeks','Month','Month','End','Month and a half','3 Days','2 Days','2 Months']
zip_terms=zip(list_terms,new_terms)
dict_terms=dict(list(zip_terms))

accounts_cleaning=accounts_cleaning.assign(Terms=accounts_cleaning['Acc_Rith'].map(dict_terms)) #creating a categories of zones

data_accounts=accounts_cleaning.drop(columns=['Acc_Plan','Acc_Rith']) #removing columns
data_accounts.duration=data_accounts.duration.astype('timedelta64[D]') #duration in days
data_accounts.head(10)
```

```python
#adding a new columns, dias_atraso
data_accounts_complementary=pd.read_csv(address+'data_accounts_complementary.csv',sep=';')
data_accounts_merge=pd.merge(data_accounts,data_accounts_complementary,left_on='Acc_ID',right_on='pec_id',how='left')
data_accounts_merge=data_accounts_merge.rename({'pec_diasatraso':'Days_impay'},axis='columns')
data_accounts=data_accounts_merge.drop(columns=['pec_id','Pec_ID'])

#formating dates.
data_accounts.Acc_Start_date=pd.to_datetime(data_accounts.Acc_Start_date, format="%Y-%m-%d")
data_accounts.Acc_Mat_date=pd.to_datetime(data_accounts.Acc_Mat_date, format="%Y-%m-%d")

data_accounts
```
